```python
# Gestion du labyrinthe
class Maze :
    
    def __init__(self,screen,graph) :
        self.graph = graph
        self.screen = screen
        self.event = Event()
        
        self.size = 0 # Epaisseur d'un chemin du labyrinthe
        self.step = 0
    
    
    # Accesseurs
    # Epaisseur / Longueur des chemins du labyrinthe 
    def set_size(self) :  
        # Choix de l'épaisseur la plus fine si l'écran n'est pas carré
        # le `0.6` est empirique
        self.size = 0.6*(self.screen.size/self.graph.size)
        
        """ Pas entre chaque node en largeur et hauteur en tenant compte
         du fait que le premier node est à self.size et le dernier self.size du bord """
        self.step = (self.screen.size - 2*self.size) / (self.graph.size-1)
    
    
    # Génération des coordonnées de chaque noeud de la liste de listes
    # `self.graph.nodes`
    def init_maze(self) :
        for nodes in self.graph.nodes :       
                for node in nodes :
                    nod_abs = float(node.value % self.graph.size)*self.step + self.size
                    nod_ord = float(node.value // self.graph.size)* self.step + self.size
                    
                    node.coords = (nod_abs, nod_ord)

                    
    # Dessin du labyrinthe parfait
    def draw_maze(self) :
        # Affichage des noeuds
        for i in range(self.graph.size) :
            for j in range(self.graph.size) :
                value = i*self.graph.size + j
                
                # Le coefficient 0.7 est empirique                                   
                circle(self.screen.my_screen, self.graph.color, \
                       self.graph.nodes[value][0].coords,0.7*self.size)
           
        
        # Affichage des liens parents/enfants
        for id_parent in range(self.graph.nb_nodes) :  # `id_parent` est le numéro du noeud
            parent = self.graph.nodes[id_parent][0]
                                                   
            for child in self.graph.nodes[id_parent] :
                id_child = child.value
                
                # Le parent est après l'enfant : rect horizontal partant de l'enfant                                   
                if id_parent - id_child == 1 :  
                    rect(self.screen.my_screen, self.graph.color, (child.coords[0]-self.size//2,  \
                            child.coords[1]-self.size//2, self.step + self.size//2, self.size))
                    
                # Le parent est avant l'enfant : rect horizontal partant du parent
                elif id_parent - id_child == -1 : 
                    rect(self.screen.my_screen, self.graph.color, (parent.coords[0]-self.size//2,  \
                            parent.coords[1]-self.size//2, self.step + self.size//2, self.size))
                    
                # Le parent est en-dessous de l'enfant : rect vertical partant de l'enfant    
                elif id_parent - id_child == self.graph.size : 
                    rect(self.screen.my_screen, self.graph.color, (child.coords[0]-self.size//2,   \
                            child.coords[1]-self.size//2, self.size, self.step + self.size//2))
                    
                # Le parent est au-dessus de l'enfant : rect vertical partant du parent    
                elif id_parent - id_child == -self.graph.size :  
                    rect(self.screen.my_screen, self.graph.color, (parent.coords[0]-self.size//2,   \
                            parent.coords[1]-self.size//2, self.size, self.step + self.size//2))
                
                # Dans les autres cas, on passe (parent sans enfant)
                else :
                    continue
               
            
    # Fin de l'application                
    def on_close(self) :
        self.event.on_close()
        
        
##################################################################################################
##################################################################################################
##################################################################################################
```
