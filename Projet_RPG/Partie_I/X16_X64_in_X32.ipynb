1/ Le script ci-dessous (<u><b>à exécuter</b></u>) permet :
- <b>d'afficher</b> l'image de base et celle transfrmée via la fonction `def show_images(img1,img2)`.
- <b>de charger l'image à transformer</b> (ici `old_firebolt.png`) et de récupérer ses dimensions (lignes 24-28). On a également le nom du nouveau fichier (ici `new_firebolt.png`).

```python
# Cette partie de script charge les bibliothèques et gère l'affichage des deux images.
from PIL import Image             # Importe la bibliothèque gérant les images.
import matplotlib.pyplot as plt   # Importe la bibliothèque permettant ici d'afficher les images.


         
# Affiche l'image de base et celle transformée
def show_images(img1,img2) :
    # Permet de créer une sous_fenêtre, le 3ème argument est la place de l'image.
    # Affichage de l'image de base.
    plt.subplot(1,2,1)
    plt.title("Image de base")
    plt.imshow(img1,cmap ='gray')
    
    # Affichage de l'image transformée.
    plt.subplot(1,2,2)
    plt.title("Image modifiée")
    plt.imshow(img2,cmap ='gray')

    

# Chargement de l'image à partir du répertoire de Jupyter.
# BIEN MONTER L'IMAGE DANS JUPYTER !
OLD_FILENAME = "old_firebolt.png"  # Nom du fichier de l'image à transformer.
NEW_FILENAME = "new_firebolt.png"  # Nom du fichier de l'image transformée.

img_base = Image.open(OLD_FILENAME) # Charge l'image à modifier.
width,height = img_base.size        # Récupère  la largeur (width) et hauteur (height) de l'image en pixels.
```
